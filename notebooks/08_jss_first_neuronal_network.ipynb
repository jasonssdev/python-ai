```python
import pandas as pd
```

```python
df_temperature = pd.read_csv(filepath_or_buffer="../data/tempeture_data.csv",sep=",")
df_temperature.head()
```

```python
X = df_temperature["celsius"].values
y = df_temperature["fahrenheit"].values
```

```python
import tensorflow as tf
```

# Model Definition

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1, )),  # input, a celsius data
    tf.keras.layers.Dense(units=1)  # output, a fahrenheit data
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss="mean_squared_error"
)
```

# Training

```python
training = model.fit(X, y, epochs=1000)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel("# Epoch")
plt.ylabel("Loss Size")
plt.plot(training.history["loss"])
```

```python
# Evaluation
```

```python


c = 100
prediction = model.predict([[c]])
print(f"Prediction: {c} celsius are {prediction[0]} fahrenheit")
```
